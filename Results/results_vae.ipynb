```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
sns.set(font_scale=1.5)
p = sns.color_palette("Set2")
sns.palplot(p)
sns.set_palette(p)
cmap = plt.cm.get_cmap('Spectral')
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 3})
```

```python
filename = 'log_vaecnn.csv'
col_names = ['epoch', 'loss', 'val_loss']
df = pd.read_csv(filename, delimiter=';', names=col_names)
df = df[df.epoch % 10 == 0]
df.head()
```

```python
fig = plt.figure(figsize=(16, 8), frameon=False)
s = sns.lineplot(data=df, x='epoch', y='loss', label='loss')
sns.lineplot(data=df, x='epoch', y='val_loss', label='val_loss')
s.set_ylabel('Loss')
s.set_xlabel('Epoch')
plt.show()
```
