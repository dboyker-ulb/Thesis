```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = plt.imread('1.png')
img2 = plt.imread('2.png')
img3 = plt.imread('3.png')
print(img1.shape)
imgs = [img1, img2, img3]
```

```python
f, axis = plt.subplots(ncols=3, nrows=1, figsize=(16, 8))
for i, ax in enumerate(axis):
    ax.axis('off')
    ax.imshow(imgs[i])
plt.savefig('plt.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
